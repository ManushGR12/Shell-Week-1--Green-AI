# 📝 Data Preprocessing for GreenAI Project

This notebook demonstrates basic data preprocessing steps for sustainable agriculture datasets.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
greenai_data = pd.read_csv('../data/GreenAI_Dataset.csv')
farmer_data = pd.read_csv('../data/farmer_advisor_dataset.csv')
market_data = pd.read_csv('../data/market_researcher_dataset.csv')

print('GreenAI Dataset shape:', greenai_data.shape)
print('Farmer Advisor Dataset shape:', farmer_data.shape)
print('Market Researcher Dataset shape:', market_data.shape)
```

```python
# Preview datasets
greenai_data.head()
```

```python
# Handle missing values
greenai_data.fillna(method='ffill', inplace=True)
farmer_data.fillna(method='ffill', inplace=True)
market_data.fillna(method='ffill', inplace=True)

print('Missing values handled successfully.')
```

```python
# Basic statistics
greenai_data.describe()
```

```python
# Visualization example
sns.histplot(greenai_data.iloc[:, 0], kde=True)
plt.title('Distribution of first column in GreenAI Dataset')
plt.show()
```
